[Test the hypothesis that the average revenue from users of the Ultimate and Surf calling plans differs.]

[Formulate the null and the alternative hypotheses, choose the statistical test, decide on the alpha value.]

```python
merged_with_plan_details['surf_sms_over_rev'] = surf_sms_over_rev
merged_with_plan_details['ultimate_sms_over_rev'] = ultimate_sms_over_rev
merged_with_plan_details['surf_data_over_rev'] = surf_data_over_rev
merged_with_plan_details['ultimate_data_over_rev'] = ultimate_data_over_rev
merged_with_plan_details['surf_minutes_over_rev'] = surf_minutes_over_rev
merged_with_plan_details['ultimate_minutes_over_rev'] = ultimate_minutes_over_rev

revenue = merged_with_plan_details[['user_id', 'monthly_cost','plan','surf_sms_over_rev', 
                                   'ultimate_sms_over_rev', 'surf_data_over_rev', 
                                   'ultimate_data_over_rev', 'surf_minutes_over_rev', 
                                   'ultimate_minutes_over_rev']].copy()

revenue['total_revenue'] = (revenue['monthly_cost'] + 
                           revenue['surf_sms_over_rev'].fillna(0) + 
                           revenue['ultimate_sms_over_rev'].fillna(0) + 
                           revenue['surf_data_over_rev'].fillna(0) + 
                           revenue['ultimate_data_over_rev'].fillna(0) + 
                           revenue['surf_minutes_over_rev'].fillna(0) + 
                           revenue['ultimate_minutes_over_rev'].fillna(0))
```

```python
surf_revenue = revenue[revenue['plan'] == 'surf']['total_revenue']
ultimate_revenue = revenue[revenue['plan'] == 'ultimate']['total_revenue']

alpha = 0.05
t_statistic, p_value = st.ttest_ind(surf_revenue, ultimate_revenue)

if p_value < alpha:
    print(f"\nDecision: Reject H0 (p-value {p_value:.4f} < alpha {alpha})")
    print("Conclusion: There IS a statistically significant difference in average revenue between plans")
    
    # Determine which plan has higher revenue
    if surf_revenue.mean() > ultimate_revenue.mean():
        print("Surf plan generates significantly higher average revenue")
    else:
        print("Ultimate plan generates significantly higher average revenue")
else:
    print(f"\nDecision: Fail to reject H0 (p-value {p_value:.4f} >= alpha {alpha})")
    print("Conclusion: There is NO statistically significant difference in average revenue between plans")

print()
print(f"Exact p-value: {p_value}")
print(f"P-value in scientific notation: {p_value: .2e}")
```
