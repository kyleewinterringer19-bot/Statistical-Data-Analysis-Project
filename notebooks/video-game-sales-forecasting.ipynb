```python
platform_totals = recent_data.sum(axis=1)

# Show platforms with their total sales
print("Platform total sales (2000-2016):")
print(platform_totals.sort_values(ascending=False))

# Filter to only platforms with sales > 0
platforms_with_sales = recent_data[platform_totals > 0]

print(f"\nPlatforms with sales: {len(platforms_with_sales)}")
print(f"Platforms filtered out: {len(recent_data) - len(platforms_with_sales)}")
```

```python
console_platforms = ['PS2', 'PS3', 'PS4', 'PS', 'X360', 'XB', 'XOne', 'Wii', 'WiiU', 'GC', 'N64', 'DC']
handheld_platforms = ['DS', '3DS', 'GBA', 'PSP', 'PSV', 'GB']
other_platforms = ['PC']

# Filter to just console platforms that have sales
console_data = platforms_with_sales.loc[
    [platform for platform in console_platforms if platform in platforms_with_sales.index]
]

print("Console platforms with sales:")
print(console_data.index.tolist())
print(f"Number of console platforms: {len(console_data)}")
```

```python
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
fig.suptitle('Console Platform Sales Trends Over Time (2000-2016)', fontsize=16)

# Flatten axes array for easier iteration
axes_flat = axes.flatten()

# Plot each console platform in its own subplot
for i, platform in enumerate(console_data.index):
    ax = axes_flat[i]
    
    # Plot the trend line for this platform
    ax.plot(console_data.columns, console_data.loc[platform], 
            marker='o', linewidth=2, markersize=4, color='darkgreen')
    
    ax.set_title(f'{platform}', fontsize=12, fontweight='bold')
    ax.set_xlabel('Year')
    ax.set_ylabel('Sales (millions)')
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
def analyze_platform_performance_by_region(data, top_n=10):
    """
    Analyze platform performance broken down by region
    Returns total sales for each platform organized by region from highest to lowest
    """
    platform_regional = data.groupby('platform')[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum()
    platform_regional['total_sales'] = platform_regional.sum(axis=1)
    platform_regional_sorted = platform_regional.sort_values(by='total_sales', ascending=False)
    top_platforms = platform_regional_sorted.head(top_n)

    sales_pivot = data.pivot_table(
        values=['na_sales', 'eu_sales', 'jp_sales', 'other_sales'],
        index='platform',
        aggfunc='sum'
    )
    
    # Create pivot table for game count by region  
    count_pivot = data.pivot_table(
        values=['na_sales', 'eu_sales', 'jp_sales', 'other_sales'],
        index='platform',
        aggfunc='count'
    )
    
    # Filter to top N platforms and sort by total sales
    top_platform_names = top_platforms.index
    sales_pivot_top = sales_pivot.loc[top_platform_names]
    count_pivot_top = count_pivot.loc[top_platform_names]
    
    return sales_pivot_top, count_pivot_top

# Function to analyze platform performance by region
```

### Cross-Regional Platform Comparison

Let's create a comparative analysis of platform performance across regions:

```python
def comparative_platform_analysis_by_region(data, start_year=2000, top_n=5):
    """
    Create comparative analysis of platform performance across regions
    Returns top platforms for each region with both sales volume and market share analysis
    """
    # Filter to recent years (last 5 years)
    recent_data = data[data['year_of_release'] >= start_year].copy()
    
    # Calculate total sales by platform for each region
    regional_analysis = {}
    regions = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
    region_names = ['North America', 'Europe', 'Japan', 'Other']
    
    for region, region_name in zip(regions, region_names):
        # Group by platform and sum sales for this region
        platform_sales = recent_data.groupby('platform')[region].sum().sort_values(ascending=False)
        
        # Calculate market share percentages
        total_regional_sales = platform_sales.sum()
        market_share = (platform_sales / total_regional_sales * 100).round(2)
        
        # Get top N platforms
        top_platforms = platform_sales.head(top_n)
        top_market_share = market_share.head(top_n)
        
        regional_analysis[region_name] = {
            'sales': top_platforms,
            'market_share': top_market_share,
            'total_sales': total_regional_sales
        }
    
    return regional_analysis

# Call the function
regional_results = comparative_platform_analysis_by_region(games, start_year=2000, top_n=5)

# Display results for each region
for region, data in regional_results.items():
    print(f"\n=== {region.upper()} ===")
    print(f"Total Regional Sales (2000-2016): {data['total_sales']:.2f} million")
    print(f"\nTop 5 Platforms by Sales Volume:")
    for i, (platform, sales) in enumerate(data['sales'].items(), 1):
        market_share = data['market_share'][platform]
        print(f"{i}. {platform}: {sales:.2f}M ({market_share}%)")# Create a comparative platform analysis
```

```python
def analyze_genre_by_region(data, genre_name, start_year=2000):
    """
    Analyze a specific genre's performance across all regions for the last 5 years
    """
    # Filter data for the specific genre and time period
    genre_data = data[(data['genre'] == genre_name) & 
                     (data['year_of_release'] >= start_year)].copy()
    
    if len(genre_data) == 0:
        print(f"No data found for {genre_name} genre in the specified period")
        return None
    
    # Regional analysis
    regional_sales = {
        'North America': genre_data['na_sales'].sum(),
        'Europe': genre_data['eu_sales'].sum(), 
        'Japan': genre_data['jp_sales'].sum(),
        'Other': genre_data['other_sales'].sum()
    }
    
    # Calculate total sales and market share
    total_sales = sum(regional_sales.values())
    market_share = {region: (sales/total_sales*100) if total_sales > 0 else 0 
                   for region, sales in regional_sales.items()}
    
    # Game count by region (all regions have same count)
    game_count = len(genre_data)
    
    # Average sales per game by region
    avg_sales = {region: sales/game_count if game_count > 0 else 0 
                for region, sales in regional_sales.items()}
    
    # Year-over-year analysis
    yearly_sales = genre_data.groupby('year_of_release')[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum()
    
    return {
        'genre': genre_name,
        'total_sales': total_sales,
        'game_count': game_count,
        'regional_sales': regional_sales,
        'market_share': market_share,
        'avg_sales_per_game': avg_sales,
        'yearly_trends': yearly_sales
    }

# Get all unique genres
all_genres = games['genre'].dropna().unique()
print(f"All genres to analyze: {sorted(all_genres)}")
print(f"Total genres: {len(all_genres)}")# Function to analyze genre performance by region
```

```python
# Convert user_score to numeric and drop NaN values
xbox_clean = filtered_games[filtered_games['platform'] == 'XOne'].copy()
pc_clean = filtered_games[filtered_games['platform'] == 'PC'].copy()

# Convert to numeric (handles 'tbd' and other non-numeric values)
xbox_clean['user_score_numeric'] = pd.to_numeric(xbox_clean['user_score'], errors='coerce')
pc_clean['user_score_numeric'] = pd.to_numeric(pc_clean['user_score'], errors='coerce')

# Drop rows with NaN user scores
xbox_clean = xbox_clean.dropna(subset=['user_score_numeric'])
pc_clean = pc_clean.dropna(subset=['user_score_numeric'])

print("After cleaning:")
print(f"Xbox One games with valid user scores: {len(xbox_clean)}")
print(f"PC games with valid user scores: {len(pc_clean)}")
```

```python
xbox_avg = xbox_clean['user_score_numeric'].mean()
pc_avg = pc_clean['user_score_numeric'].mean()

print(f"\nAverage user ratings:")
print(f"Xbox One: {xbox_avg:.2f}")
print(f"PC: {pc_avg:.2f}")
print(f"Difference: {abs(xbox_avg - pc_avg):.2f}")
```

```python
# Extract the clean user score data
xbox_scores = xbox_clean['user_score_numeric']
pc_scores = pc_clean['user_score_numeric']

# Perform Welch's t-test (assumes unequal variances)
from scipy import stats
t_statistic, p_value = stats.ttest_ind(xbox_scores, pc_scores, equal_var=False)

print("Hypothesis Test Results:")
print(f"Sample sizes: Xbox One = {len(xbox_scores)}, PC = {len(pc_scores)}")
print(f"Sample means: Xbox One = {xbox_scores.mean():.3f}, PC = {pc_scores.mean():.3f}")
```

```python
print(f"t-statistic: {t_statistic:.4f}")
print(f"p-value: {p_value:.6f}")
print(f"Alpha level: 0.05")

# Interpret the results
if p_value < 0.05:
    print("Result: Reject the null hypothesis")
    print("Conclusion: Xbox One and PC platforms have significantly different average user ratings")
else:
    print("Result: Fail to reject the null hypothesis") 
    print("Conclusion: No significant difference between Xbox One and PC platform ratings")
```
